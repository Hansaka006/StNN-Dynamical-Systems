```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the loss data from the two CSV files
file1 = "ANN_loss_curve_new_val.csv"
file2 = "SNN_loss_curve_p_6_new_vali.csv"

# Assuming the CSV files have columns 'Step' and 'Loss'
loss1 = pd.read_csv(file1)
loss2 = pd.read_csv(file2)

# Find the minimum MSE values
min_mse_ann = loss1['Loss'].min()
min_mse_snn = loss2['Loss'].min()

# Plot the loss curves
plt.figure(figsize=(5, 3))
plt.plot(loss1['Step'], loss1['Loss'], label="FFNN validation MSE", linestyle='-', color='b')
# Draw horizontal lines at the minimum MSE values
plt.axhline(y=min_mse_ann, color='b', linestyle=':', label=f'Min FFNN MSE: {min_mse_ann:.2e}')


plt.plot(loss2['Step'], loss2['Loss'], label="StNN validation MSE", linestyle='--', color='r')
plt.axhline(y=min_mse_snn, color='r', linestyle=':', label=f'Min StNN MSE: {min_mse_snn:.2e}')

# Add plot details
#plt.title("Comparison of Training Loss Curves")
plt.xlabel("Steps")
plt.ylabel("MSE")
plt.yscale("log")  # Optional: Use logarithmic scale for loss
plt.legend()
plt.grid(True)

# Save the plot to a PDF file
output_file = "loss_comparison_val.pdf"
plt.savefig(output_file, format="pdf", bbox_inches="tight")
#plt.close()  # Close the plot to free memory

print(f"Plot saved as '{output_file}'.")
# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# File paths for the four CSV files
csv_files = [
    "cumulative_mean_error_p_1.csv",
    "cumulative_mean_error_p_2.csv",
    "cumulative_mean_error_p_4.csv",
    "cumulative_mean_error_p_6.csv",
]

# Labels for each curve (corresponding to different p-values)
p_values = ["p=1", "p=2", "p=4", "p=6"]

# Initialize the plot
plt.figure(figsize=(5, 3))

# Loop through each CSV file and plot its data
for csv_file, p_label in zip(csv_files, p_values):
    data = pd.read_csv(csv_file)
    plt.plot(data["Time Step"], data["Cumulative Mean Error"], label=p_label)

# Add title, labels, and legend
#plt.title("Cumulative Mean Error for Different p Values")
plt.xlabel("Time Step")
plt.ylabel("Cumulative MSE")
plt.legend()
#plt.grid()

# Save the combined plot to a PDF file
output_file = "cumulative_mean_error_comparison.pdf"
plt.savefig(output_file, format="pdf", bbox_inches="tight")
plt.tight_layout()
plt.show()

print(f"Plot saved to {output_file}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.size'] = 15
# Load data from CSV files
x_data = pd.read_csv('Havok_predict_x.csv')
y_data = pd.read_csv('Havok_predict_y.csv')
z_data = pd.read_csv('Havok_predict_z.csv')

# Extract time and values
time = x_data['Time']
true_x, predicted_x = x_data['Truth'], x_data['Reconstruction']
true_y, predicted_y = y_data['Truth'], y_data['Reconstruction']
true_z, predicted_z = z_data['Truth'], z_data['Reconstruction']

# Compute MSE
mse_x = np.mean((true_x - predicted_x) ** 2)
mse_y = np.mean((true_y - predicted_y) ** 2)
mse_z = np.mean((true_z - predicted_z) ** 2)

# Compute mean MSE
mean_mse = np.mean([mse_x, mse_y, mse_z])

print(f"MSE (x): {mse_x:.6f}")
print(f"MSE (y): {mse_y:.6f}")
print(f"MSE (z): {mse_z:.6f}")
print(f"Mean MSE: {mean_mse:.6f}")

# Create the plot
plt.figure(figsize=(12, 6))

# Plot x data
plt.subplot(3, 1, 1)
plt.plot(time, true_x, c="b", label="Ground Truth")
plt.plot(time, predicted_x, "--", c="r", label="Predicted")
plt.legend()
plt.ylabel("x")

# Plot y data
plt.subplot(3, 1, 2)
plt.plot(time, true_y, c="b")
plt.plot(time, predicted_y, "--", c="r")
plt.ylabel("y")

# Plot z data
plt.subplot(3, 1, 3)
plt.plot(time, true_z, c="b")
plt.plot(time, predicted_z, "--", c="r")
plt.ylabel("z")
plt.xlabel("Time")

# Save and show the plot
output_file = "predicted_Havok.pdf"
plt.savefig(output_file, format="pdf", bbox_inches="tight")
plt.show()
```
